```python
import pandas as pd
import numpy as np
```

```python
monfils = pd.read_csv('/Users/jasontang/Desktop/cogs108/Group053_SP25/monfils_serve_chart.csv')
monfils.drop(columns=['Unnamed: 0'], inplace=True)
monfils = monfils.replace('-', np.NaN)
```

```python
string_cols=[ 'Unret%', '<=3 W%',
       'RiP W%', 'RiP W%.1', 'SvImpact', '1st: Unret%', '1st: <=3 W%',
       'RiP W%.2', 'RiP W%.3', 'SvImpact1st', 'D Wide%', 'A Wide%', 'BP Wide%',
       '2nd: Unret%', '2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    monfils[col] = monfils[col].str.replace('%', '')
    monfils[col] = monfils[col].astype('float64')
monfils    
```
