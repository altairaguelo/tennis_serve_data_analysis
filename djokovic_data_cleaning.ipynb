```python
import pandas as pd
import numpy as np
```

```python
djo = pd.read_csv('djokovic_serve_chart.csv')
djo.drop(columns=['Unnamed: 0'], inplace=True)
djo = djo.replace('-', np.nan)
```

```python
djo
```

```python
string_cols=[ 'Unret%', '<=3 W%',
       'RiP W%', 'RiP W%.1', 'SvImpact', '1st: Unret%', '1st: <=3 W%',
       'RiP W%.2', 'RiP W%.3', 'SvImpact1st', 'D Wide%', 'A Wide%', 'BP Wide%',
       '2nd: Unret%', '2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    djo[col] = djo[col].str.replace('%', '')
    djo[col] = djo[col].astype('float64')
djo    
```
