```python
import pandas as pd
import numpy as np
```

```python
fed = pd.read_csv('../data/federer_serve_chart.csv')
fed = fed.replace('-', np.NaN)
fed.head()
```

```python
fed.dtypes #show data types to see which cols to convert to numeric
```

```python
fed.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    fed[col] = fed[col].str.replace('%', '')
    fed[col] = fed[col].astype('float64')
fed.head()  
```

```python
fed['tourney'].unique()
```

```python
surfaces = {'Wimbledon': 'grass',
           'Roland Garros': 'clay',
           'Doha': 'hard',
           'Australian Open': 'hard',
           'Tour Finals': 'hard',
           'Basel': 'hard',
           'Shanghai': 'hard',
           'Laver Cup': 'hard',
           'US Open': 'hard',
            'Cincinnati': 'hard',
           'Halle': 'grass',
           'Rome': 'clay',
           'Madrid': 'clay',
           'Miami': 'hard',
           'Indian Wells': 'hard',
           'Dubai': 'hard',
           'Paris': 'hard',
           'Stuttgart': 'grass',
           'Rotterdam': 'hard',
           'Canada': 'hard',
           'Monte Carlo': 'clay',
           'Brisbane': 'hard',
           'Istanbul': 'hard',
           'Davis Cup': 'hard',
           'Hamburg': 'clay',
           'London Olympics': 'grass',
           'Stockholm': 'hard',
           'Masters Cup': 'hard',
           'Beijing Olympics': 'hard',
           'Estoril': 'clay',
           'Tokyo': 'hard',
           'Bangkok': 'hard',
           'Gstaad': 'clay',
           'Vienna': 'hard',
           'Munich': 'clay',
           'Marseille': 'hard',
           'Milan': 'clay',
           'Sydney': 'hard',
           'Sydney Olympics': 'hard',
           's Hertogenbosch': 'grass'}
fed['surface'] = fed['tourney'].apply(lambda x: surfaces[x])
fed.head()
```

```python
fed = pd.get_dummies(fed, columns=['surface', 'result'])
fed.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
fed.columns
```

```python
fed.to_csv('../data/federer_serve_chart.csv', index=False)
```

```python
fed.isna().sum() #show the number of missing values in each column
```

```python
#use simple imputation to fill in missing values with the column mean
fed_imputed = fed.copy()
fed_imputed['1st: A BP Wide%'].fillna(fed_imputed['1st: A BP Wide%'].mean(), inplace=True)
fed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
fed_imputed.to_csv('../data/federer_serve_chart_imputed.csv', index=False)
```

```python
fed_speed = pd.read_csv('../data/federer_serve_speed.csv')
fed_speed.head()
```

```python
#one hot encoding for surface and result
fed_speed['surface'] = fed_speed['tourney'].apply(lambda x: surfaces[x])
fed_speed = pd.get_dummies(fed_speed, columns=['surface', 'result'])
fed_speed.drop(columns=['result_L'], inplace=True)
fed_speed.head()
```

```python
# save the data
fed_speed.to_csv('../data/federer_serve_speed.csv', index=False)
```

```python
fed_speed.isna().sum() #show the number of missing values in each column
```

```python
# use simple imputation to fill in missing values with the column mean
fed_speed_imputed = fed_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: max','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    fed_speed_imputed[col].fillna(fed_speed_imputed[col].mean(), inplace=True)
fed_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
fed_imputed.to_csv('../data/federer_serve_chart_imputed.csv', index=False)
```
