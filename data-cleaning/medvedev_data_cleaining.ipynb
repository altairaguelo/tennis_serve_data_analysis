```python
import pandas as pd
import numpy as np
```

```python
medvedev = pd.read_csv('../data/medvedev_serve_chart.csv')
medvedev = medvedev.replace('-', np.nan)
medvedev.head()
```

```python
medvedev.dtypes #show data types to see which cols to convert to numeric
```

```python
medvedev.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    medvedev[col] = medvedev[col].str.replace('%', '')
    medvedev[col] = medvedev[col].astype('float64')
medvedev.head()  
```

```python
medvedev = pd.get_dummies(medvedev, columns=['surface', 'result'])
medvedev.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
medvedev.columns
```

```python
medvedev.to_csv('../data/medvedev_serve_chart.csv', index=False)
```

```python
#use simple imputation to fill in missing values with the column mean
medvedev_imputed = medvedev.copy()
medvedev_imputed['1st: A BP Wide%'].fillna(medvedev_imputed['1st: A BP Wide%'].mean(), inplace=True)
medvedev_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
medvedev_speed = pd.read_csv('../data/medvedev_serve_speed.csv')
medvedev_speed.head()
```

```python
#one hot encoding for surface and result
medvedev_speed['surface'] = medvedev_speed['tourney'].apply(lambda x: surfaces[x])
medvedev_speed = pd.get_dummies(medvedev_speed, columns=['surface', 'result'])
medvedev_speed.drop(columns=['result_L'], inplace=True)
medvedev_speed.head()
```

```python
# save the data
medvedev_speed.to_csv('../data/medvedev_serve_speed.csv', index=False)
```

```python
medvedev_speed.isna().sum() #show the number of missing values in each column
```

```python
# use simple imputation to fill in missing values with the column mean
medvedev_speed_imputed = medvedev_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: max','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    medvedev_speed_imputed[col].fillna(medvedev_speed_imputed[col].mean(), inplace=True)
medvedev_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
medvedev_speed_imputed.to_csv('../data/medvedev_serve_speed_imputed.csv', index=False)
```
