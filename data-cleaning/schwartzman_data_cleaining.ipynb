```python
import pandas as pd
import numpy as np
```

```python
man = pd.read_csv('../data/schwartzman_serve_chart.csv')
man = man.replace('-', np.nan)
man.head()
```

```python
man.dtypes #show data types to see which cols to convert to numeric
```

```python
man.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    man[col] = man[col].str.replace('%', '')
    man[col] = man[col].astype('float64')
man.head()  
```

```python
man['tourney'].unique()
```

```python
surfaces = {'Wimbledon': 'grass',
           'Roland Garros': 'clay',
           'Doha': 'hard',
           'Australian Open': 'hard',
           'Tour Finals': 'hard',
           'Basel': 'hard',
           'Shanghai': 'hard',
           'Laver Cup': 'hard',
           'US Open': 'hard',
           'Cincinnati': 'hard',
           'Halle': 'grass',
           'Rome': 'clay',
           'Madrid': 'clay',
           'Miami': 'hard',
           'Indian Wells': 'hard',
           'Dubai': 'hard',
           'Paris': 'hard',
           'Stuttgart': 'grass',
           'Rotterdam': 'hard',
           'Canada': 'hard',
           'Monte Carlo': 'clay',
           'Brisbane': 'hard',
           'Istanbul': 'hard',
           'Davis Cup': 'hard',
           'Hamburg': 'clay',
           'London Olympics': 'grass',
           'Stockholm': 'hard',
           'Masters Cup': 'hard',
           'Beijing Olympics': 'hard',
           'Estoril': 'clay',
           'Tokyo': 'hard',
           'Bangkok': 'hard',
           'Gstaad': 'clay',
           'Vienna': 'hard',
           'Munich': 'clay',
           'Marseille': 'hard',
           'Milan': 'clay',
           'Sydney': 'hard',
           'Sydney Olympics': 'hard',
           's Hertogenbosch': 'grass',
           'Buenos Aires': 'clay',
           'Barcelona': 'clay',
           "Queen's Club": 'grass',
           'Rio de Janeiro': 'clay',
           'Atp Cup': 'hard',
           'Antwerp': 'hard',
           'Cologne 2': 'hard',
           'Cordoba': 'clay',
           'Los Cabos': 'hard',
           'Acapulco': 'hard',
           'Marrakech': 'clay',
           'Sao Paulo': 'clay',
           'Montevideo CH': 'clay',
           'Guayaquil CH': 'clay',
           'Corrientes CH': 'clay',
           'Potro Alegre CH': 'clay',
           'Campinas CH': 'clay',
           'Bastad': 'clay'}
man['surface'] = man['tourney'].apply(lambda x: surfaces[x])
man.head()
```

```python
man = pd.get_dummies(man, columns=['surface', 'result'])
man.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
man.columns
```

```python
man.to_csv('../data/schwartzman_serve_chart.csv', index=False)
```

```python
man.isna().sum() #show the number of missing values in each column
```

```python
#use simple imputation to fill in missing values with the column mean
man_imputed = man.copy()
man_imputed['1st: A BP Wide%'].fillna(man_imputed['1st: A BP Wide%'].mean(), inplace=True)
man_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
man_imputed.to_csv('../data/schwartzman_serve_chart_imputed.csv', index=False)
```

```python
man_speed = pd.read_csv('../data/schwartzman_serve_speed.csv')
man_speed.head()
```

```python
#one hot encoding for surface and result
man_speed['surface'] = man_speed['tourney'].apply(lambda x: surfaces[x])
man_speed = pd.get_dummies(man_speed, columns=['surface', 'result'])
man_speed.drop(columns=['result_L'], inplace=True)
man_speed.head()
```

```python
# save the data
man_speed.to_csv('../data/schwartzman_serve_speed.csv', index=False)
```

```python
man_speed.isna().sum() #show the number of missing values in each column
```

```python
# use simple imputation to fill in missing values with the column mean
man_speed_imputed = man_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: max','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    man_speed_imputed[col].fillna(man_speed_imputed[col].mean(), inplace=True)
man_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
man_speed_imputed.to_csv('../data/schwartzman_serve_speed_imputed.csv', index=False)
```
