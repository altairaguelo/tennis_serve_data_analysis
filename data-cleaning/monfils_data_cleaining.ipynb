```python
import pandas as pd
import numpy as np
```

```python
monfils = pd.read_csv('../data/monfils_serve_chart.csv')
monfils = monfils.replace('-', np.NaN)
monfils.head()
```

```python
monfils.dtypes #show data types to see which cols to convert to numeric
```

```python
monfils.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    monfils[col] = monfils[col].str.replace('%', '')
    monfils[col] = monfils[col].astype('float64')
monfils.head()  
```

```python
monfils[monfils['tourney'] == 'Stuttgart']
```

```python
surfaces = {'Miami': 'hard',
           'Indian Wells': 'hard',
           'Dubai': 'hard',
           'Australian Open': 'hard',
           'Auckland': 'hard',
           'Brisbane': 'hard',
           'Shanghai': 'hard',
           'Beijing': 'hard',
           'US Open': 'hard',
            'Cincinnati': 'hard',
           'Canada': 'hard',
           'Wimbledon': 'grass',
           'Roland Garros': 'clay',
           'Monte Carlo': 'clay',
           'Estoril': 'clay',
           'Doha': 'hard',
           'Stockholm': 'hard',
           'Madrid': 'clay',
           'Adelaide 1': 'hard',
           'Vienna': 'hard',
           'Sofia': 'hard',
           'Rotterdam': 'hard',
           'Montpellier': 'hard',
           'Paris': 'clay',
           'Antwerp': 'hard',
           'Kaohsiung CH': 'hard',
           'Eastbourne': 'grass',
           'Tour Finals': 'hard',
           'Tokyo': 'hard',
           'Washington': 'hard',
           'Davis Cup': 'clay',
           'Stuttgart': 'grass',
           'Marseille': 'hard',
           'Halle': 'grass',
           'Beijing Olympics': 'hard',
           'Rome': 'clay'}
monfils['surface'] = monfils['tourney'].apply(lambda x: surfaces[x])
monfils.head()
```

```python
monfils = pd.get_dummies(monfils, columns=['surface', 'result'])
monfils.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
monfils.columns
```

```python
monfils.to_csv('../data/monfils_serve_chart.csv', index=False)
```

```python
monfils.isna().sum() #show the number of missing values in each column
```

```python
#use simple imputation to fill in missing values with the column mean
monfils_imputed = monfils.copy()
monfils_imputed['1st: A BP Wide%'].fillna(monfils_imputed['1st: A BP Wide%'].mean(), inplace=True)
monfils_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
monfils_imputed.to_csv('../data/monfils_serve_chart_imputed.csv', index=False)
```

```python
monfils_speed = pd.read_csv('../data/monfils_serve_speed.csv')
monfils_speed.head()
```

```python
#one hot encoding for surface and result
monfils_speed['surface'] = monfils_speed['tourney'].apply(lambda x: surfaces[x])
monfils_speed = pd.get_dummies(monfils_speed, columns=['surface', 'result'])
monfils_speed.drop(columns=['result_L'], inplace=True)
monfils_speed.head()
```

```python
# save the data
monfils_speed.to_csv('../data/monfils_serve_speed.csv', index=False)
```

```python
monfils_speed.isna().sum() #show the number of missing values in each column
```

```python
# use simple imputation to fill in missing values with the column mean
monfils_speed_imputed = monfils_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: max','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    monfils_speed_imputed[col].fillna(monfils_speed_imputed[col].mean(), inplace=True)
monfils_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
monfils_imputed.to_csv('../data/monfils_serve_chart_imputed.csv', index=False)
```
