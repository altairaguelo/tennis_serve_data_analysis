```python
import pandas as pd
import numpy as np
```

```python
isner = pd.read_csv('isner_serve_chart.csv')
isner.drop(columns=['Unnamed: 0'], inplace=True)
isner = isner.replace('-', np.nan)
```

```python
string_cols=[ 'Unret%', '<=3 W%',
       'RiP W%', 'RiP W%.1', 'SvImpact', '1st: Unret%', '1st: <=3 W%',
       'RiP W%.2', 'RiP W%.3', 'SvImpact1st', 'D Wide%', 'A Wide%', 'BP Wide%',
       '2nd: Unret%', '2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    isner[col] = isner[col].str.replace('%', '')
    isner[col] = isner[col].astype('float64')
isner    
```
