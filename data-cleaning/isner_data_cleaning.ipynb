```python
import pandas as pd
import numpy as np
```

```python
isner = pd.read_csv('../data/isner_serve_chart.csv')
isner = isner.replace('-', np.nan)
isner.head()
```

```python
isner.dtypes #show data types to see which cols to convert to numeric
```

```python
isner.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    isner[col] = isner[col].str.replace('%', '')
    isner[col] = isner[col].astype('float64')
isner.head()  
```

```python
isner['tourney'].unique()
```

```python
surfaces = {'Wimbledon': 'grass',
           'Roland Garros': 'clay',
           'Doha': 'hard',
           'Australian Open': 'hard',
           'Tour Finals': 'hard',
           'Basel': 'hard',
           'Shanghai': 'hard',
           'Laver Cup': 'hard',
           'US Open': 'hard',
            'Cincinnati': 'hard',
           'Halle': 'grass',
           'Rome': 'clay',
           'Madrid': 'clay',
           'Miami': 'hard',
           'Indian Wells': 'hard',
           'Dubai': 'hard',
           'Paris': 'hard',
           'Stuttgart': 'grass',
           'Rotterdam': 'hard',
           'Canada': 'hard',
           'Monte': 'clay',
           'Brisbane': 'hard',
           'Istanbul': 'hard',
           'Davis': 'hard',
           'Hamburg': 'clay',
           'London Olympics': 'grass',
           'Stockholm': 'hard',
           'Masters': 'hard',
           'Beijing': 'hard',
           'Estoril': 'clay',
           'Tokyo': 'hard',
           'Bangkok': 'hard',
           'Gstaad': 'clay',
           'Vienna': 'hard',
           'Munich': 'clay',
           'Marseille': 'hard',
           'Milan': 'clay',
           'Sydney': 'hard',
           's': 'grass',
           'Houston': 'clay',
           'Dallas': 'hard',
           'Newport': 'grass',
           'Atlanta': 'hard',
           'Acapulco': 'hard',
           'Washington': 'hard',
           'New York': 'hard',
           'Auckland': 'hard',
           "Queen's Club": 'grass',
           'Nice': 'clay'}
isner['surface'] = isner['tourney'].apply(lambda x: surfaces[x])
isner.head()
```

```python
isner = pd.get_dummies(isner, columns=['surface', 'result'])
isner.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
isner.columns
```

```python
isner.to_csv('../data/isner_serve_chart.csv', index=False)
```

```python
isner.isna().sum() #show the number of missing values in each column
```

```python
#use simple imputation to fill in missing values with the column mean
isner_imputed = isner.copy()
isner_imputed['1st: A BP Wide%'].fillna(isner_imputed['1st: A BP Wide%'].mean(), inplace=True)
isner_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
isner_imputed.to_csv('../data/isner_serve_chart_imputed.csv', index=False)
```

```python
isner_speed = pd.read_csv('../data/isner_serve_speed.csv')
isner_speed.head()
```

```python
#one hot encoding for surface and result
isner_speed['surface'] = isner_speed['tourney'].apply(lambda x: surfaces[x])
isner_speed = pd.get_dummies(isner_speed, columns=['surface', 'result'])
isner_speed.drop(columns=['result_L'], inplace=True)
isner_speed.head()
```

```python
# save the data
isner_speed.to_csv('../data/isner_serve_speed.csv', index=False)
```

```python
isner_speed.isna().sum() #show the number of missing values in each column
```

```python
# use simple imputation to fill in missing values with the column mean
isner_speed_imputed = isner_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: max','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    isner_speed_imputed[col].fillna(isner_speed_imputed[col].mean(), inplace=True)
isner_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
isner_speed_imputed.to_csv('../data/isner_serve_speed_imputed.csv', index=False)
```
