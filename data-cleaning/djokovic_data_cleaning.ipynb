```python
import pandas as pd
import numpy as np
```

```python
djo = pd.read_csv('../data/djokovic_serve_chart.csv')
djo = djo.replace('-', np.nan)
djo.head()
```

```python
djo.dtypes #show data types to see which cols to convert to numeric
```

```python
djo.columns
```

```python
string_cols=[ 'overall: Unret%','overall: <=3 W%', 'overall: RiP W%',
             'overall: SvImpact','1st: Unret%', '1st: <=3 W%', 
             '1st: RiP W%', '1st: SvImpact','1st: D Wide%', '1st: A Wide%', 
             '1st: A BP Wide%', '2nd: Unret%','2nd: <=3 W%', '2nd: RiP W%']

for col in string_cols:
    djo[col] = djo[col].str.replace('%', '')
    djo[col] = djo[col].astype('float64')
djo.head()  
```

```python
djo = pd.get_dummies(djo, columns=['surface', 'result'])
djo.drop(columns=['result_L'], inplace=True)
#columns after one hot encoding
djo.columns
```

```python
djo.to_csv('../data/djokovic_serve_chart.csv', index=False)
```

```python
djo.isna().sum() #show the number of missing values in each column
```

```python
#use simple imputation to fill in missing values with the column mean
djo_imputed = djo.copy()
djo_imputed['1st: A BP Wide%'].fillna(djo_imputed['1st: A BP Wide%'].mean(), inplace=True)
djo_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
#save the imputed data
djo_imputed.to_csv('../data/djokovic_serve_chart_imputed.csv', index=False)
```

```python
djo_speed = pd.read_csv('../data/djokovic_serve_speed.csv')
djo_speed.head()
```

```python
#one hot encoding for surface and result
djo_speed['surface'] = djo_speed['tourney'].apply(lambda x: surfaces[x])
djo_speed = pd.get_dummies(djo_speed, columns=['surface', 'result'])
djo_speed.drop(columns=['result_L'], inplace=True)
djo_speed.head()
```

```python
djo_speed.to_csv('../data/djokovic_serve_speed.csv', index=False)
```

```python
djo_speed.isnull().sum()
```

```python
# use simple imputation to fill in missing values with the column mean
djo_speed_imputed = djo_speed.copy()
cols_to_impute = ['1st: avg','1st: stdev','1st: t avg','1st: wide avg','1st: min','2nd: avg','2nd: stdev',
                  '2nd: t avg','2nd: wide avg','2nd: max']
for col in cols_to_impute:
    djo_speed_imputed[col].fillna(djo_speed_imputed[col].mean(), inplace=True)
djo_speed_imputed.isna().sum() #show the number of missing values in each column after imputation
```

```python
djo_speed_imputed.to_csv('../data/djokovic_serve_speed_imputed.csv', index=False)
```
