# Research Question

What is the relationship between serve and match outcome in tennis(specifically point and match outcomes, aka points won within a set and match win/loss)? Does this reflect in the player's win percentage?
Some possible measures include first serve/second serve, placement of serve(wide, t, body), kind of point(regular, break), surface type, serve speed, etc. We will be looking specifically at men's pro tennis data to answer these questions.



## Background and Prior Work

Our group is interested in the possible relationship between serve and match outcome in tennis. To conduct our analysis, we first want to be able to distinguish what makes a “good serve” by measuring aspects of the serve such as placement, speed, type, etc. If there is a measurable distinction, we hope to discover those variables and their significance to winning a point. Furthermore, through these measures we hope to see if the serve and match outcome relationship is reflected in a player’s win percentage, so we can see how important the serve is to a player’s game.

We are not the first people to analyze the tennis serve, and its effects on match outcomes; however, there are some differences in their work when compared to the analyses that we want to pursue. For example, Abhinav Maddineni, Samarth Gowda, Eric Liu, and Daniel Liang performed analyses on ATP data to find out what factors (if any) affect serves, if the type of court affects a match’s outcome, and what metrics are important for analyzing the competitiveness of a tennis game.<a name="cite_ref-1"></a>[<sup>1</sup>](#cite_note-1) Notably, their work focused on what factors affect the serve, like player height or court surface, and they made comparisons between winners and losers of matches by looking at various statistics. Our approach is fundamentally different because the data we have available looks at individual player statistics over the course of their entire career. Because the dataset is more granular, we hope to show how the serve affects individual success. Additionally, we seek to focus specifically on the serve and how it relates to a player's match outcomes and win percentage. 

In another article, we found that analyses have been performed on the probability of winning a point in men's elite tennis where the researchers found that the first serve is crucial to increase the probability of winning the point as well as the length of the rally.<a name="cite_ref-2"></a>[<sup>2</sup>](#cite_note-2) The main difference between our problem and the article’s is that the article has a general focus on what comes after the serve rather than a complete focus on the performance of the serve.
```python
plot = sns.boxplot(data=df, x='result_W', y='1st: Unret%')
x_labels = ['Loss', 'Win']
plt.xlabel('result_W (Win/Loss Match Result)')
plot.set_xticklabels(x_labels)
```

This box plot shows that on average, when players won matches had a greater ace percentage. With outliers of those who had really "good" serving games, those who had "good" serving games but still ended up losing the match, and those who had "bad" serving matches but still end up winning.

```python
from scipy.stats import ttest_ind, chisquare, normaltest

t, p = ttest_ind(df[df['result_W'] == True]['1st: Unret%'], 
           df[df['result_W'] == False]['1st: Unret%'])
print(p) # p-value is 0.0001, so we reject the null hypothesis that there is no difference in 1st serve unret% between wins and losses
            # in other words, aces do make a difference in match outcomes
```

We found that the p-value is 0.0001, so we reject the null hypothesis that there is no difference in 1st serve unret% between wins and losses. In other words, aces do make a difference in match outcomes!

```python
#stratisfied k-fold cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=4)

thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# Logistic Regression with L1 penalty
clf_l1 = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)
# using StratifiedKFold for cross-validation because of class imbalance

for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_l1.fit(X_train, Y_train)
    y_probs_l1 = clf_l1.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_l1 = (y_probs_l1 >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_l1))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_l1))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")
```

We used stratified $5$-fold cross validation to train the model and we use a range of threshold values to find the best threshold in terms of accuracy and F1-score. The thresholds found come out to $0.5$ for both cases, and this is the same as the default value in the `predict` method.

```python
final_pred_l1 = clf_l1.predict(X_test) # didn't use threshold since the default is 0.5 which is the same as the best threshold for accuracy and f1 score
print("Classification Report for L1 Penalty Model: \n", classification_report(Y_test, final_pred_l1))
cm_l1 = confusion_matrix(Y_test, clf_l1.predict(X_test))
plt.figure(figsize=(12, 5))
sns.heatmap(cm_l1, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for L1 Penalty Model')
```

The model achieves a good level of performance as seen in the high precision and recall scores when predicting wins. We also see a good accuracy at $84%$. It seems that the model struggles a little bit more when predicting losses, however.

```python
# try model on Schwartzman data
schwartzman_pred_l1 = clf_l1.predict(schwartzman_X)
print("Classification Report for L1 Penalty Model on Schwartzman Data: \n", classification_report(schwartzman_Y, schwartzman_pred_l1))

cm_schwartzman_l1 = confusion_matrix(schwartzman_Y, schwartzman_pred_l1)
plt.figure(figsize=(12, 5))
sns.heatmap(cm_schwartzman_l1, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for L1 Penalty Model on Schwartzman Data')
```

The Schwartzman data throws our L1 model's performance off a little bit. We see a noticeably lower accuracy score at $72%$. Interestingly, we see that the loss class recall is a little bit higher, but overall, the model is not as good at predicting a less successful player's outcomes.

```python
thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# Logistic Regression with L2 penalty
clf_l2 = LogisticRegression(max_iter=3000) #raise max_iter to ensure convergence
# using StratifiedKFold for cross-validation because of class imbalance

for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_l2.fit(X_train, Y_train)
    y_probs_l2 = clf_l2.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_l2 = (y_probs_l2 >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_l2))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_l2))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")

#final model output for L2 penalty
print("Classification Report for L2 Penalty Model: \n", classification_report(Y_test, clf_l2.predict(X_test)))
cm_l2 = confusion_matrix(Y_test, clf_l2.predict(X_test))
plt.figure(figsize=(12, 5))
sns.heatmap(cm_l2, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for L2 Penalty Model')
```

We try the same method of finding the best threshold values and get $0.5$ again. We see that L2 performs a little bit worse on accuracy as well as the precision for both classes and the recall for the win class. The results are very close, however, and the only difference is we gained two false negatives. L2 performs a little bit worse.

```python
# svm classification
thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# SVM with linear kernel
clf_svm = svm.SVC(kernel='linear', probability=True, random_state=42)

# using StratifiedKFold for cross-validation because of class imbalance
for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_svm.fit(X_train, Y_train)
    y_probs_svm = clf_svm.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_svm = (y_probs_svm >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_svm))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_svm))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")
```

The code above uses 5 fold stratified cross validation to find the best threshold for win/loss classification for the svm model. The thresholds found when using F1-scores and accuracy scores are the same and happen to be the default thresholds of $0.5$.
