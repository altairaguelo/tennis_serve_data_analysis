# Research Question

What is the relationship between serve and match outcome in tennis(specifically point and match outcomes, aka points won within a set and match win/loss)? Does this reflect in the player's win percentage?
Some possible measures include first serve/second serve, placement of serve(wide, t, body), kind of point(regular, break), surface type, serve speed, etc. We will be looking specifically at men's pro tennis data to answer these questions.



## Background and Prior Work

Our group is interested in the possible relationship between serve and match outcome in tennis. Furthermore, through these measures we hope to see if the serve and match outcome relationship is reflected in a player’s win percentage, so we can see how important the serve is to a player’s game.

```python
from scipy.stats import ttest_ind, chisquare, normaltest

t, p = ttest_ind(df[df['result_W'] == True]['1st: Unret%'], 
           df[df['result_W'] == False]['1st: Unret%'])
print(p) # p-value is 0.0001, so we reject the null hypothesis that there is no difference in 1st serve unret% between wins and losses
            # in other words, aces do make a difference in match outcomes
```

We found that the p-value is 0.0001, so we reject the null hypothesis that there is no difference in 1st serve unret% between wins and losses. In other words, aces do make a difference in match outcomes!

## Classification Models: Predicting Match Outcomes
To set our baselines, first we will create logistic regression models, an SVM model, and random forest models. From these initial models, we can begin to view relationships and the validity of our datasets. We can also look back at them as we create better models and explore relationships in our data further. By doing this, we hope to see which features of the serve are important in determining match outcomes.

```python
# trying out classification in preparation for the next part of the project
from sklearn.linear_model import  LogisticRegression
from sklearn.model_selection import train_test_split
X = df.iloc[:, 5:-1]
Y = df['result_W']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
clf_l1 = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)
clf_l2 = LogisticRegression(max_iter=1000) # default is l2 penalty and lbfgs solver
clf_l1.fit(X_train, Y_train)
clf_l2.fit(X_train, Y_train)
print("L1 penalty coefficients:")
print(clf_l1.coef_)
print("L2 penalty coefficients:")
print(clf_l2.coef_)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
# Evaluate the models
y_pred_l1 = clf_l1.predict(X_test)
y_pred_l2 = clf_l2.predict(X_test)
print("L1 Penalty Model Accuracy:", accuracy_score(Y_test, y_pred_l1))
print("L2 Penalty Model Accuracy:", accuracy_score(Y_test, y_pred_l2))
print("L1 Penalty Model Classification Report:\n", classification_report(Y_test, y_pred_l1))
print("L2 Penalty Model Classification Report:\n", classification_report(Y_test, y_pred_l2))
# Confusion matrices
cm_l1 = confusion_matrix(Y_test, y_pred_l1)
cm_l2 = confusion_matrix(Y_test, y_pred_l2)
plt.figure(figsize=(12, 5))
sns.heatmap(cm_l1, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for L1 Penalty Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
from sklearn.metrics import roc_curve, auc
# Calculate ROC curve and AUC for L1 penalty model
fpr_l1, tpr_l1, _ = roc_curve(Y_test, clf_l1.predict_proba(X_test)[:, 1])
roc_auc_l1 = auc(fpr_l1, tpr_l1)
# Calculate ROC curve and AUC for L2 penalty model
fpr_l2, tpr_l2, _ = roc_curve(Y_test, clf_l2.predict_proba(X_test)[:, 1])
roc_auc_l2 = auc(fpr_l2, tpr_l2)
# Plot ROC curves
plt.figure(figsize=(10, 6))
plt.plot(fpr_l1, tpr_l1, color='blue', label='L1 Penalty (AUC = {:.2f})'.format(roc_auc_l1))
plt.plot(fpr_l2, tpr_l2, color='orange', label='L2 Penalty (AUC = {:.2f})'.format(roc_auc_l2))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line
plt.title('ROC Curve for L1 and L2 Penalty Models')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
```

```python
from sklearn import svm
clf_svm = svm.SVC(kernel='linear', C=1.0, random_state=42)
clf_svm.fit(X_train, Y_train)
y_pred_svm = clf_svm.predict(X_test)
print("SVM Model Accuracy:", accuracy_score(Y_test, y_pred_svm))
print("SVM Model Classification Report:\n", classification_report(Y_test, y_pred_svm))
cm_svm = confusion_matrix(Y_test, y_pred_svm)
plt.figure(figsize=(12, 5))
sns.heatmap(cm_svm, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for SVM Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

```python
# roc cruve for SVM
fpr_svm, tpr_svm, _ = roc_curve(Y_test, clf_svm.decision_function(X_test))
roc_auc_svm = auc(fpr_svm, tpr_svm)
# Plot ROC curve for SVM
plt.figure(figsize=(10, 6))
plt.plot(fpr_svm, tpr_svm, color='green', label='SVM (AUC = {:.2f})'.format(roc_auc_svm))
plt.plot([0, 1], [0, 1], color='gray', linestyle='--')  # Diagonal line
plt.title('ROC Curve for SVM Model')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint

rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, Y_train)
y_pred_rf = rf.predict(X_test)
print("Random Forest Model Accuracy:", accuracy_score(Y_test, y_pred_rf))
print("Random Forest Model Classification Report:\n", classification_report(Y_test, y_pred_rf))
cm_rf = confusion_matrix(Y_test, y_pred_rf)
plt.figure(figsize=(12, 5))
sns.heatmap(cm_rf, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for Random Forest Model')
plt.xlabel('Predicted')
plt.ylabel('Actual')
```

## Now cross validate model to check if we can do better!

```python
X = df.iloc[:, 5:-1] # features
Y = df['result_W'] # classification target

X_cv, X_test, Y_cv, Y_test = train_test_split(X, Y, test_size=0.2, random_state=4)

#stratified k-fold cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=4)

thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# Logistic Regression with L1 penalty
clf_l1 = LogisticRegression(penalty='l1', solver='liblinear', max_iter=1000)
# using StratifiedKFold for cross-validation because of class imbalance

for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_l1.fit(X_train, Y_train)
    y_probs_l1 = clf_l1.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_l1 = (y_probs_l1 >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_l1))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_l1))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")
```

Here, we see that the f1 threshold and accuracy threshold end up being the same as the default of the model anyway, but we wanted to explore it for the sake of completeness.

```python
# try model on Nadal data
nadal_pred_l1 = clf_l1.predict(nadal_X)
print("Classification Report for L1 Penalty Model on nadal Data: \n", classification_report(nadal_Y, nadal_pred_l1))

cm_nadal_l1 = confusion_matrix(nadal_Y, nadal_pred_l1)
plt.figure(figsize=(12, 5))
sns.heatmap(cm_nadal_l1, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
```

After cross validation, we see that our model performs worse at predicting Nadal's match outcome/win/losses. However, this may be an indicator that our model may be generalizing better, and instead is more accurate for the whole of men's ATP tennis players, rather than just the greatest players.

```python
thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# Logistic Regression with L2 penalty
clf_l2 = LogisticRegression(max_iter=3000) #raise max_iter to ensure convergence
# using StratifiedKFold for cross-validation because of class imbalance

for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_l2.fit(X_train, Y_train)
    y_probs_l2 = clf_l2.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_l2 = (y_probs_l2 >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_l2))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_l2))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")

print("Classification Report for L2 Penalty Model: \n", classification_report(Y_test, clf_l2.predict(X_test)))
cm_l2 = confusion_matrix(Y_test, clf_l2.predict(X_test))
plt.figure(figsize=(12, 5))
sns.heatmap(cm_l2, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
```

```python
thresholds = np.arange(0.1, 1.0, 0.1)
threshold_f1_scores = {t: [] for t in thresholds}
threshold_accuracy_scores = {t: [] for t in thresholds}

# Logistic Regression with L2 penalty
clf_l2 = LogisticRegression(max_iter=3000) #raise max_iter to ensure convergence
# using StratifiedKFold for cross-validation because of class imbalance

for train_index, test_index in skf.split(X_cv, Y_cv):
    X_train, X_validation = X_cv.iloc[train_index], X_cv.iloc[test_index]
    Y_train, Y_validation = Y_cv.iloc[train_index], Y_cv.iloc[test_index]
    clf_l2.fit(X_train, Y_train)
    y_probs_l2 = clf_l2.predict_proba(X_validation)[:, 1]
    for threshold in thresholds:
        y_pred_l2 = (y_probs_l2 >= threshold).astype(int)
        threshold_f1_scores[threshold].append(f1_score(Y_validation, y_pred_l2))
        threshold_accuracy_scores[threshold].append(accuracy_score(Y_validation, y_pred_l2))

# Calculate average F1 score and accuracy
avg_f1_cv = {t: np.mean(scores) for t, scores in threshold_f1_scores.items()}
avg_accuracy_cv = {t: np.mean(scores) for t, scores in threshold_accuracy_scores.items()}

best_threshold_f1 = max(avg_f1_cv, key=avg_f1_cv.get)
best_threshold_accuracy = max(avg_accuracy_cv, key=avg_accuracy_cv.get)
print(f"Best f1 threshold: {best_threshold_f1}")
print(f"Best accuracy threshold: {best_threshold_accuracy}")

#final model output for L2 penalty
print("Classification Report for L2 Penalty Model: \n", classification_report(Y_test, clf_l2.predict(X_test)))
cm_l2 = confusion_matrix(Y_test, clf_l2.predict(X_test))
plt.figure(figsize=(12, 5))
sns.heatmap(cm_l2, annot=True, fmt='d', cmap='Blues', xticklabels=['Loss', 'Win'], yticklabels=['Loss', 'Win'])
plt.title('Confusion Matrix for L2 Penalty Model')
```

We try the same method of finding the best threshold values and get $0.5$ again. We see that L2 performs a little bit worse on accuracy as well as the precision for both classes and the recall for the win class. The results are very close, however, and the only difference is we gained two false negatives. L2 performs a little bit worse.

# Discussion and Conclusion

The purpose of this report is to see if we can find a relationship between a player’s serve and the outcome of the match. Initially hypothesizing that a strong serve will lead to a favorable outcome, but also noting that other variables such as type of serve, placement of the serve, speed, surface, and type of point are equally as likely to affect the match outcome. Our data consists of match statistics of 5 successful professional men’s players with an emphasis on their serve and the speed. We looked at stats such as outcome, surface, the percentage of unreturned serves, the win percentage of points that took less than 3 hits (serve, return, winning hit) or less, serve impact (a calculation of how many points were won because of the serve), serve aggression, and win percentage when the serve was returned, for both first and second serves. For speed, we looked at the speed for the different placements a serve can go, average serve speed, standard deviation, and minimum and maximum speeds, all for each type of serve, match outcome and surface. For speed, we also had to impute data, as it was significantly less, but it would lead to spikes in values since a lot of missing values were filled with the averages. We looked at these in a concatenated set with data from each player, as well as singling out one player, Roger Federer.

We noted interesting findings, for example, due to the fact that our players were successful, we found that the dataset was heavily skewed with wins, having 1265 wins and 534 losses for outcomes. With Federer accounting for 487 wins and 164 losses, which is around 37% of our matches, which can also have a skew on the data. This can be attributed to the fact that our data is charted by fans, and he is a fan favorite to watch, but also that he has the most matches played overall so more will be charted. More specifically, on the serve, we noticed that for first serves, our players' serves were unreturned 40-45% of the time, and 50-60% of the time, they would win the point in under 3 hits. Which is impressive, as it shows that our players are getting nearly half of their points from either a winner off the return or just the serve alone. While compared to 2nd serves, unreturned is around 10-20% and 20-30%, which would be normal since they are mainly less risky serves, but also that a failure to hit it in results in a loss of the point. What was interesting to look at was that the win rate for our players when their first serves returned around 50-60% and their second returned 40-50%. Which shows that although players do have strong serves, getting the serve in is the most important. Which is shown as second serves also have a negative serve aggression while they are still able to win 40-50% of the points off of it. We also noticed a change in surface. The impact and speed on hard courts were relatively the same, clay had worse impact and slower speeds, and grass had a higher impact and faster speeds. Showcasing that a stronger server would do better on grass compared to if they played on grass or clay. While also looking at speed, we noticed that our graphs showed a lot of outliers and had a thicker interquartile range, but that could be due to the need to impute our data. We also saw that faster serves have more losses; this can be an indication of match loss or that players that serve faster are not winning. As well as that, those who are losing tend to take on more risk in their serve in order to win points back. Lastly, comparing results and 1st serve aces, we found a p-value of 0.0001, rejecting the null hypothesis that there is no difference, showcasing that they do make a difference in match outcomes.

Using our data, we trained models to compare the serve statistics and the overall result of the match. Looking at models of logistic regression, we were able to achieve an accuracy of 83%. For SVMs, we found that it was correct 82% of the time, and for Random Forest, it was right 83% of the time. We chose these since logistic regression is best for predicting the probability of a binary outcome in which we have wins and losses. SVM is good for drawing a hard line in the classification and creating wins and losses. Random Forest was used to look at more complex parameters and see improved accuracy. It is interesting to note that the feature with most importance for the random forest was overall win percentage when the ball was returned. As these are points won off a hit that wasn’t the serve getting it in more of an importance. Since we noted that our dataset had significantly more wins than losses, we believed that our models would work best on successful players. To test this, we used the data of two more players, one who was more successful, Rafa Nadal, whose data has a record of 361 wins and 136 losses. And a rather more average player, Diego Schwartzman, whose data has a record of 57 wins and 63 losses. We found that our models for Schwartzman had a worse accuracy than our dataset of around 72%. Whilst Nadal had a similar accuracy of around 83%. Which could be a sign that our models are just better at classifying wins and being correct since we have more, instead of overall success.

Overall we concluded that a stronger serve will lead to more wins. As well as that, stronger servers will fare better on grass courts, worse on clay courts, and will be more neutral on hard courts. The first serve will be the best serve to win points off of, however it is more important for the second serve to be hit in rather than to win off of. There are limitations to this, since our data was trained off of male professional players, it could be less or more effective for female professional players and less experienced players. As well, since our data was extracted from a website that was fan-charted, statistics from every match of a player's career are not present. This is significant, as instead of a sample size of around 5000 matches, we were just limited to around 1800 and even less for speed data. Another limitation is that most players chosen started their careers in tennis more than a decade ago; many new players coming in known for their serves have very little data we could use, as well as a change in tennis rackets and form that could affect the effectiveness of a serve. Our work could have some negative and positive outcomes for society, as it can be used in a negative way, potentially affecting a player's bioinformation and bonuses based on predictions and data. It could also be used for gambling to hit player propositions on the number of aces hit by a player. It also shows those with less experience in tennis that the serve is an important aspect of your game. Not only is a strong first serve a good way to win points, but also a very consistent second serve is important in order to be successful.
